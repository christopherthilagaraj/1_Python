```python
#Print Even Number in the List
a=(20,10,16,19,25,1,276,188)
for i in a:
    if (i%2)==0:
        print(i,"is Even")
```

```python
#Print Odd Number in the List
a=(20,10,16,19,25,1,276,188)
for i in a:
    if (i%2)!=0:
        print(i,"is Odd")
```
