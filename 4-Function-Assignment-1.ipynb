```python
#Create a function that Checks whether the given number is Odd or Even
```

```python
def OddEven():
    num=int(input("Enter a number:-"))
    if (num%2)==0:
        print(num," is Even number")
    else:
        print(num," is Odd number")    
        
```

```python
#Create a function that tells elegibility of marriage for male and female according to age limit like 21 for male and 18 for femail
```

```python
def MarriageElegible(Gender,Age):
    if Gender=="Male":
        if Age>=22:
            print("Gender= Male, Age=",Age,"Your are Eligible to marriage")
        else:
            print("Gender= Male, Age=",Age,"Your are not Eligible to marriage")
    else:
        if Age>=18:
            print("Gender= Female, Age=",Age,"Your are Eligible to marriage")
        else:
            print("Gender= Female, Age=",Age,"Your are not Eligible to marriage")     
  
```

```python
MarriageElegible("Female",22)
```

```python
#Calculate the percentage of your 10th mark
```

```python
def CalculateMark(subject1,subject2,subject3,subject4,subject5):
    total=subject1+subject2+subject3+subject4+subject5
    per=total/500*100
    return per
```

```python
CalculateMark(50,30,50,60,80)
```

```python
#print area and perimeter of triangle using class and functions
def AreaofTriangle(Height,Breadth):
    AreaofTriangle=(Height*Breadth)/2
    print("Area formula:(Height*Breadth)/2")
    return AreaofTriangle    
```

```python
result=AreaofTriangle(32,34)
print("Area of Triangle:",result)
```
